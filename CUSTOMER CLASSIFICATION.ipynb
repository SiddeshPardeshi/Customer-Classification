```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
cust_train['age_band'] = le.fit_transform(cust_train['age_band'])
cust_train['family_income'] = le.fit_transform(cust_train['family_income'])
cust_train['status'] = le.fit_transform(cust_train['status'])
cust_train['occupation'] = le.fit_transform(cust_train['occupation'])
cust_train['occupation_partner'] = le.fit_transform(cust_train['occupation_partner'])
cust_train['home_status'] = le.fit_transform(cust_train['home_status'])
cust_train['gender'] = le.fit_transform(cust_train['gender'])
cust_train['region'] = le.fit_transform(cust_train['region'])
cust_train['TVarea'] = le.fit_transform(cust_train['TVarea'])
```

```python
#using replace function 
cust_train['self_employed'].replace(to_replace = 'Yes', value = 1, inplace = True)
cust_train['self_employed'].replace(to_replace = 'No', value = 0, inplace = True)
```

```python
cust_train['self_employed_partner'].replace(to_replace = 'Yes', value = 1, inplace = True)
cust_train['self_employed_partner'].replace(to_replace = 'No', value = 0, inplace = True)
```

```python
cust_train.drop(['post_code','post_area'],axis=1,inplace=True)  #dropping post code and post area
```

```python
for col in cust_train.iloc[:,1:29].columns:
    print(col)
    sns.boxplot(x = cust_train[col], data = cust_train)
    plt.show()
```

```python
def Outlier(var):
    for x in var.iloc[:, 1:29].columns:
        Q1 = var[x].quantile(0.25)
        Q3 = var[x].quantile(0.75)
        IQR = Q3-Q1
        Lower = Q1-(1.5*IQR)
        Upper = Q3+(1.5*IQR)
        var.loc[:,x] = np.where(var[x].values > Upper, Upper, var[x].values)
        var.loc[:,x] = np.where(var[x].values < Lower, Lower, var[x].values)
    return var
df_train = Outlier(cust_train)
```

```python
for col in cust_train.iloc[:, 1:29].columns:
    print(col)
    sns.boxplot(x = cust_train[col], data = cust_train)
    plt.show()
```

```python
x = df_train.iloc[:,1:29]
y = df_train['Revenue_Grid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(class_weight = 'balanced', max_iter = 10000)
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred_test = lr.predict(x_test)
```

```python
y_pred_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
def model_eval(actual, predicted):
    cm = confusion_matrix(actual, predicted)
    accuracy = accuracy_score(actual, predicted)
    class_rep = classification_report(actual, predicted)
    print('The cm of the Model is :', cm)
    print('The accuracy of the Model is :', accuracy)
    print(class_rep)
```

```python
model_eval(y_test, y_pred_test)   #checking for test data
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(y_test, y_pred_test)
```
